#1. What is the difference between a function and a method in Python?
- **Function**: A block of reusable code that is defined outside of a class. It can be called directly.


```python
#Example:

def greet(name):
    return f"Hello, {name}!"
print(greet("Ankit"))
```

- **Method**: A function that is associated with an object and is defined inside a class. It is called using the object or class.

```python
#Example:

class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"
obj = Greeter()
print(obj.greet("Ankit"))
```

#6. Explain the concept of generators in Python and how they are defined.
- Generators are special functions that yield values one at a time using the yield keyword.

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1
for num in count_up_to(3):
    print(num)
```

#7. What are the advantages of using generators over regular functions?

- Generators are memory-efficient since they produce items one at a time.
- Useful for handling large datasets or infinite sequences.

```python
#Example:

def generate_numbers(n):
    for i in range(n):
        yield i
# Memory-efficient iteration
for num in generate_numbers(1000000):
    if num > 5:
        break
    print(num)
```

#10. What is the difference between map(), reduce(), and filter() functions in Python?

- map(): Applies a function to each element of an iterable.




```python
#Example:

nums = [1, 2, 3]
print(list(map(lambda x: x * 2, nums)))
```

- filter(): Filters elements of an iterable based on a condition.


```python
#Example:

nums = [1, 2, 3, 4]
print(list(filter(lambda x: x % 2 == 0, nums)))
```

- reduce(): Reduces an iterable to a single value using a function.


```python
#Example:

from functools import reduce
nums = [1, 2, 3, 4]
print(reduce(lambda x, y: x + y, nums))
```

# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.




```python
def sum_of_evens(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Taking user input
user_input = input("Enter a list of numbers separated by spaces: ")
numbers = [int(num) for num in user_input.split()]

print("Sum of even numbers:", sum_of_evens(numbers))
```

#4. Write a Python function that checks if a given number is prime or not from 1 to 200.




```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Taking user input for the range
start = int(input("Enter the starting number: "))
end = int(input("Enter the ending number: "))

# Finding prime numbers in the given range
primes = [num for num in range(start, end + 1) if is_prime(num)]


print(f"Prime numbers between {start} and {end}: {primes}")
```

#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.




```python
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        self.count += 1
        fib = self.a
        self.a, self.b = self.b, self.a + self.b
        return fib

# Example
fib_iter = Fibonacci(10)
for num in fib_iter:
    print(num, end=" ")
```

#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.



```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Taking user input for the exponent
exponent = int(input("Enter the exponent: "))

# Example usage
print(f"Powers of 2 up to 2^{exponent}:")
for power in powers_of_two(exponent):
    print(power)
```

#7. Implement a generator function that reads a file line by line and yields each line as a string.



```python
def read_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

# Example usage
# if file "example.txt" exists
for line in read_lines("example.txt"):
    print(line)
```

#9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.



```python
def celsius_to_fahrenheit(celsius):
    return list(map(lambda c: (c * 9/5) + 32, celsius))

# Taking user input
user_input = input("Enter temperatures in Celsius separated by spaces: ")

# Convert the input string to a list of integers
temps_celsius = [float(temp) for temp in user_input.split()]


print("Temperatures in Fahrenheit:", celsius_to_fahrenheit(temps_celsius))
```

#10. Create a Python program that uses filter() to remove all the vowels from a given string.

```python
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda char: char not in vowels, s))

# Taking user input
user_input = input("Enter a string: ")

print("String after removing vowels:", remove_vowels(user_input))
```

#Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.

- Write a Python program using lambda and map.

```python
# Input data
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

# Function to calculate order value and add 10€ if necessary
def calculate_order(order):
    order_number = order[0]
    quantity = order[2]
    price_per_item = order[3]
    total_value = quantity * price_per_item
    if total_value < 10000:  # Apply the 10€ bonus for orders less than 10,000€
        total_value += 10
    return (order_number, total_value)

# Using map with lambda function
result = list(map(lambda order: (order[0], (order[2] * order[3]) + 10) if order[2] * order[3] < 10000 else (order[0], order[2] * order[3]), orders))


print(result)
```
